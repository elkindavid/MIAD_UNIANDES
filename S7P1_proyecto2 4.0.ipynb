## Datos para la predicción de género en películas

![image info](https://raw.githubusercontent.com/albahnsen/MIAD_ML_and_NLP/main/images/moviegenre.png)

En este proyecto se usará un conjunto de datos de géneros de películas. Cada observación contiene el título de una película, su año de lanzamiento, la sinopsis o plot de la película (resumen de la trama) y los géneros a los que pertenece (una película puede pertenercer a más de un género). Por ejemplo:
- Título: 'How to Be a Serial Killer'
- Plot: 'A serial killer decides to teach the secrets of his satisfying career to a video store clerk.'
- Generos: 'Comedy', 'Crime', 'Horror'

La idea es que usen estos datos para predecir la probabilidad de que una película pertenezca, dada la sinopsis, a cada uno de los géneros.

```python
# function for text cleaning 
def clean_text(text):
    # remove backslash-apostrophe 
    text = re.sub("\'", "", text) 
    # remove everything except alphabets 
    text = re.sub("[^a-zA-Z]"," ",text) 
    # remove whitespaces 
    text = ' '.join(text.split()) 
    # convert text to lowercase 
    text = text.lower() 
    
    return text
```

```python
def freq_words(x, terms = 30): 
  all_words = ' '.join([text for text in x]) 
  all_words = all_words.split() 
  fdist = nltk.FreqDist(all_words) 
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())}) 
  
  # selecting top 20 most frequent words 
  d = words_df.nlargest(columns="count", n = terms) 
  
  # visualize words and frequencies
  plt.figure(figsize=(6,6)) 
  ax = sns.barplot(data=d, x= "count", y = "word") 
  ax.set(ylabel = 'Word') 
  plt.show()
```

```python
dataTraining['clean_plot'] = dataTraining['plot'].apply(lambda x: clean_text(x))
dataTesting['clean_plot'] = dataTesting['plot'].apply(lambda x: clean_text(x))
```

```python
# print 30 most frequent words 
freq_words(dataTraining['clean_plot'], 30)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# function to remove stopwords
def remove_stopwords(text):
    no_stopword_text = [w for w in text.split() if not w in stop_words]
    return ' '.join(no_stopword_text)

dataTraining['clean_plot'] = dataTraining['clean_plot'].apply(lambda x: remove_stopwords(x))
dataTesting['clean_plot'] = dataTesting['clean_plot'].apply(lambda x: remove_stopwords(x))
```

```python
# Definición de variable de interés (y)
dataTraining['genres'] = dataTraining['genres'].map(lambda x: eval(x))
le = MultiLabelBinarizer()
y = le.fit_transform(dataTraining['genres'])

tfidf_vectorizer = TfidfVectorizer(max_df=0.2, sublinear_tf=True, smooth_idf=False, max_features=40000)

# split dataset into training and validation set
xtrain, xval, ytrain, yval = train_test_split(dataTraining['clean_plot'], y, test_size=0.2, random_state=9)

# create TF-IDF features
xtrain_tfidf = tfidf_vectorizer.fit_transform(xtrain)
xval_tfidf = tfidf_vectorizer.transform(xval)
```

```python
lr = LogisticRegression(solver='liblinear', penalty='l2')
clf = OneVsRestClassifier(lr)

# fit model on train data
clf.fit(xtrain_tfidf, ytrain)

# predict probabilities
y_pred_prob = clf.predict_proba(xval_tfidf)

t = 0.3 # threshold value
y_pred_new = (y_pred_prob >= t).astype(int)

# evaluate performance
f1_score(yval, y_pred_new, average="micro")

# # Impresión del desempeño del modelo
roc_auc_score(yval, y_pred_prob, average='macro')
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import CountVectorizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC
import tensorflow as tf

# Preprocesar los datos
X = dataTraining['clean_plot']
y = dataTraining['genres']

# Convertir los géneros en variables binarias
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(y)

# Crear el vectorizador de palabras
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X)

# Convertir la matriz dispersa a un array numpy denso
X_train = X_train.toarray()

# Convertir y_train a un array numpy
y_train = np.array(y)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Crear el modelo de la red neuronal
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))

# Crear la métrica AUC
auc_metric = AUC()

# Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy', auc_metric])

# Entrenar el modelo
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluar el modelo
loss, accuracy, auc_score = model.evaluate(X_test, y_test)
print('Loss:', loss)
print('Accuracy:', accuracy)
print('AUC:', auc_score)
```

```python
# Random Search para Red Neuronal
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.feature_extraction.text import CountVectorizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC
import tensorflow as tf
from sklearn.metrics import f1_score, roc_auc_score

# Preprocesar los datos
X = dataTraining['clean_plot']
y = dataTraining['genres']

# Convertir los géneros en variables binarias
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(y)

# Crear el vectorizador de palabras
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X)

# Convertir la matriz dispersa a un array numpy denso
X_train = X_train.toarray()

# Convertir y_train a un array numpy
y_train = np.array(y)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Función para crear el modelo de la red neuronal
def create_model(learning_rate=0.001):
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
    model.add(Dropout(0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(y.shape[1], activation='softmax'))
    
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=learning_rate), metrics=['accuracy', AUC()])
    return model

# Crear el modelo de KerasClassifier para usar con RandomizedSearchCV
model = KerasClassifier(build_fn=create_model)

# Definir los parámetros e hiperparámetros a buscar
param_grid = {
    'learning_rate': [0.001, 0.01, 0.1],
    'batch_size': [16, 32, 64],
    'epochs': [10, 20, 30]
}

# Realizar la búsqueda aleatoria con validación cruzada
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, cv=3, scoring='roc_auc', n_iter=10, random_state=42)
random_search_result = random_search.fit(X_train, y_train)

# Obtener los mejores hiperparámetros
best_params = random_search_result.best_params_

# Obtener el modelo con los mejores hiperparámetros
best_model = random_search_result.best_estimator_

# Evaluar el modelo con los mejores hiperparámetros en los datos de prueba
y_pred = best_model.predict(X_test)
f1 = f1_score(y_test, y_pred, average='micro')
auc = roc_auc_score(y_test, y_pred, average='micro')

# Mostrar los resultados
print("Mejores hiperparámetros:", best_params)
print("F1 Score:", f1)
print("AUC Score:", auc)
```

### Make model available with Flask

```python
# Importación librerías
from flask import Flask, render_template
from flask_restx import Api, Resource, fields
```

```python
# Definición aplicación Flask
app = Flask(__name__)

# Definición API Flask
api = Api(
    app, 
    version='1.0', 
    title='Movie genre classification API',
    description='Movie genre classification API')

ns = api.namespace('Predict', 
     description='Top 3 Most Likely Movie Genres')

# Definición argumentos o parámetros de la API
parser = api.parser()

parser.add_argument(
    'PLOT', 
    type=str, 
    required=True, 
    help='Plot', 
    location='args')

resource_fields = api.model('Resource', {
    'result': fields.String
})
```

```python
# Definición de la clase para disponibilización
@ns.route('/')
class MovieGenreApi(Resource):

    @api.doc(parser=parser)
    @api.marshal_with(resource_fields)
    def get(self):
        args = parser.parse_args()
        
        return {
         "result": predict(args['PLOT'])
        }, 200
```

```python
# Ejecución de la aplicación que disponibiliza el modelo de manera local en el puerto 5000
app.run(debug=True, use_reloader=False, host='0.0.0.0', port=5000)
```
